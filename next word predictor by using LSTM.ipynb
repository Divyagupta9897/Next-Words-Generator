```python
#now we assign numbers to the text by tokenizer class
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
len(tokenizer.word_index)
```

```python
#now we will try to make datasets by sentences to sentences
for sentence in faqs.split('\n'):
    print(sentence)
```

```python
# now we convert these sentences into numbers
for sentence in faqs.split('\n'):
    print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
#now we make input and ouput combo
input_sequence=[]
for sentence in faqs.split('\n'):
    tokenizer_sentence=tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1,len( tokenizer_sentence)):
         input_sequence.append(tokenizer_sentence[:i+1])
    
```

```python
input_sequence
```

```python
# here we can see that the size of every array or list is not equals so we have to use zero padding so firstly we have to check maximum length of sentence 
max_len=max([len(x) for x in input_sequence])
```

```python
#zero padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequence,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences # here we can see that we got input and output both
```

```python
x=padded_input_sequences[:,:-1]
```

```python
x # input
```

```python
y=padded_input_sequences[:,-1]
```

```python
y #output
```

```python
print(x.shape)
print(y.shape)
```

```python
# here we do multiple  class classification for the output
#so now we transfor our output by onehotencodding
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=283) # 283 is len of words
```

```python
y.shape
```

```python
y
```

```python
#now we will prepare the architecture of lstm after that we will do trainning
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM , Dense, Embedding,InputLayer
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(56,)))
model.add(Embedding(283,100))
model.add(LSTM(150))
model.add(Dense(283,activation='softmax'))
```

```python
#now we do compile
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from sklearn.model_selection import train_test_split

# Assuming x and y are your input and output data
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)

# Train the model and calculate validation accuracy
history = model.fit(x_train, y_train, epochs=100, validation_data=(x_val, y_val))

# To check the training and validation accuracy after training:
print(history.history.keys())  # Lists all metrics tracked, including accuracy and val_accuracy

# You can also plot the validation accuracy:
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
